Does the Feynman technique improve the exam scores?

c1 : Feynman technique
c2 : exam scores

c1 is the predictor and c2 the predicted variable.

H0: there is no significant difference between students who used Feynman technique and users who used classical study methods.
H1: The students using the Feynman technique perform better in the exam compared to students using classical study methods.

Since the two groups are disjointed, the study is between subjects. I will evaluate the results by using an independent t-test.

The normality test was performed on both groups to assess if the data was normally distributed. The normality test computed on the group using the classical study method (STD: 5.007, M: 85.933) obtained a p-value of **0.775**, while the Feynman technique group (STD: 3.976, M: 91.667) obtained a p-value of **0.614**. Since both the values are above the confidence threshold set to 0.05 I can conclude that the samples of both groups are normally distributed.
This is confirmed by a visual inspection of the distribution (using kde interpolation) that didn't reveal any significant outliers. 

The independent t-test returned a p-value of **0.002** which is clearly below the threshold 0.05. **t(28)=-3.473**. I can state with confidence that the null hypothesis H0 can be rejected and consequentially accept H1. The improvement can be noticed by comparing the two means reporting a difference of **5.733** points among the two approaches.

```python
import pandas as pd 
import numpy as np
from scipy import stats
import seaborn 

df = pd.DataFrame(data={"g1": np.array([78, 85, 92, 88, 95, 80, 83, 90, 87, 82, 89, 91, 79, 84, 86]), "g2": np.array([88, 95, 91, 98, 93, 89, 94, 97, 85, 92, 96, 90, 88, 93, 86])})

normal1Stat, normal1P = stats.normaltest(df["g1"])
normal2Stat, normal2P = stats.normaltest(df["g2"])

seaborn.displot(df[["g1", "g2"]], kind="kde")


print(f"[G1]\n\
        Normality test of p-value: {normal1P:.3f}\n\
        Mean: {df['g1'].mean():.3f}\n\
        Stddev: {df['g1'].std():.3f}")

print(f"[G2]\n\
        Normality test of p-value: {normal2P:.3f}\n\
        Mean: {df['g2'].mean():.3f}\n\
        Stddev: {df['g2'].std():.3f}")


tt = stats.ttest_ind(df["g1"], df["g2"])

print(f"[Independent T-Test]\n\
        t({tt.df:.0f})={tt.statistic:.3f}\n\
        p-value: {tt.pvalue:.3f}")

print(f"Mean difference: {df['g2'].mean() - df['g1'].mean() : .3f}")
```
